Q1 Dataset Exploration
Perform the following exploratory steps:
Display the first 10 rows
Display the dataset structure using info()
Find the number of rows and columns
Identify the data types of each column

```python
import pandas as pd
df = pd.read_csv("Advanced_pandas_sales_dataset.csv")

# Display first 10 rows
print(df.head(10))

# Dataset structure
print(df.info())

# Number of rows and columns
print("Shape:", df.shape)

# Data types of each column
print(df.dtypes)
```

Q2 Sales Filtering Analysis
Filter the dataset to display:
Orders where Sales > 3500
Orders where Region = "East" and Sales > 2500
Explain what insights you observe.

```python
# 1. Orders where Sales > 3500
high_sales = df[df['Sales'] > 3500]
print("Orders with Sales > 3500:")
print(high_sales)

# 2. Orders where Region = "East" AND Sales > 2500
east_high = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]
print("\nEast region orders with Sales > 2500:")
print(east_high)

# Insights
print("\nInsight: High-value orders in East region indicate"
      " strong purchasing power or premium product demand.")
```

Q3 Customer Purchase Analysis
Identify the top 5 customers based on total sales.
Hint: Use GroupBy with aggregation

```python
# Group by Customer, sum Sales, get top 5
top5_customers = (df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(5))

print("Top 5 Customers by Total Sales:")
print(top5_customers)
```

Q4 Regional Sales Performance
Calculate the total sales generated by each region.
Then identify:
Which region generated the highest revenue?

```python
# Total sales by region
regional_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("Total Sales by Region:")
print(regional_sales)

# Region with highest revenue
top_region = regional_sales.idxmax()
print(f"\nHighest Revenue Region: {top_region}")
print(f"Total Sales: {regional_sales[top_region]:.2f}")
```

Q5 Product Performance Analysis
Calculate the total quantity sold for each product.
Which product appears to be the most popular?

```python
# Total quantity sold per product
product_qty = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
print("Total Quantity Sold by Product:")
print(product_qty)

# Most popular product
print(f"\nMost Popular Product: {product_qty.idxmax()}")
print(f"Units Sold: {product_qty.max()}")
```

Q6 Multi-Level Grouping
Group the dataset by: Region and Category
Calculate the total sales for each combination.
What insights can you derive from this analysis?

```python
# Group by Region AND Category, calculate total sales
multi_group = (df.groupby(['Region', 'Category'])['Sales'].sum().reset_index().rename(columns={'Sales': 'Total Sales'}))

print(multi_group)

# Insight
best_combo = multi_group.loc[multi_group['Total Sales'].idxmax()]
print(f"\nBest Region-Category combo: {best_combo['Region']} - {best_combo['Category']}")
print(f"Total Sales: {best_combo['Total Sales']:.2f}")
```

Q7 Pivot Table Creation
Create a pivot table showing:
Rows → Region
Columns → Product
Values → Total Sales
Which region sells the most laptops?


```python
# Pivot table: Rows=Region, Columns=Product, Values=Total Sales
pivot = df.pivot_table(
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum',
    fill_value=0
)

print(pivot)
# Which region sells the most laptops?
if 'Laptop' in pivot.columns:
    top_laptop_region = pivot['Laptop'].idxmax()
    print(f"\nRegion with most Laptop sales: {top_laptop_region}")
```

Q8 Profitability Analysis
Calculate the average profit for each product category.
Which category is most profitable?

```python
# Average profit per product category
avg_profit = (df.groupby('Category')['Profit'].mean().sort_values(ascending=False))

print("Average Profit by Category:")
print(avg_profit.round(2))

# Most profitable category
print(f"\nMost Profitable Category: {avg_profit.idxmax()}")
print(f"Avg Profit: {avg_profit.max():.2f}")
```

Q9 Discount Impact Study
Analyze the relationship between discount and profit.
Tasks:
Calculate the average profit for each discount level
Identify whether higher discounts lead to lower profit

```python
# Average profit per discount level
discount_profit = (df.groupby('Discount')['Profit'].mean().sort_index())

print("Average Profit by Discount Level:")
print(discount_profit.round(2))

# Check correlation
corr = df['Discount'].corr(df['Profit'])
print(f"\nCorrelation between Discount and Profit: {corr:.4f}")
print("Insight: Negative correlation means higher discounts → lower profit."
       if corr < 0 else "Insight: Discounts do not reduce profit significantly.")
```

Q10 End-to-End Data Wrangling Task
Perform the following workflow:
Filter the dataset where Sales > 3000
Group data by Region and Product
Calculate total sales
Reshape the result into a pivot table
Explain the business insights obtained from this analysis.

```python
# Step 1: Filter Sales > 3000
filtered = df[df['Sales'] > 3000]

# Step 2 & 3: Group by Region + Product, calculate total sales
grouped = (filtered.groupby(['Region', 'Product'])['Sales'].sum().reset_index().rename(columns={'Sales': 'Total Sales'}))

print("Grouped Result (Sales > 3000):")
print(grouped)

# Step 4: Reshape into pivot table
final_pivot = grouped.pivot_table(
    values='Total Sales',
    index='Region',
    columns='Product',
    aggfunc='sum',
    fill_value=0
)

print("\nFinal Pivot Table:")
print(final_pivot)

# Business Insights
top_r = final_pivot.sum(axis=1).idxmax()
top_p = final_pivot.sum(axis=0).idxmax()
print(f"\nBusiness Insight:")
print(f"  Top Region (high-value orders): {top_r}")
print(f"  Top Product (high-value orders): {top_p}")
print("  Focus marketing efforts on this region-product combo.")
```
