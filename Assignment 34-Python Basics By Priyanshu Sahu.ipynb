## Ques 7 Create a list a five numbers and print the maximum number from the list

```python
lst=[11,12,13,14,15]
Maximum_number=lst[0]
index=0
for i in range(len(lst)):
    if lst[i]>Maximum_number:
        Maximum_number=lst[i]
        index=i
print(f"your Max number is {Maximum_number} at index {index}")
```

## Question 10 Write a Python program to count how many times a number appears in a list.
