Q1. Creating a DataFrame
Create a Pandas DataFrame using a Python dictionary with the following columns:
. Product
. Region
. Sales
. Quantity

```python
import pandas as pd

data = {
    'Product'  : ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'Region'   : ['East',   'West',  'East',   'West',    'East'],
    'Sales'    : [1500, 1200, 800, 2000, 600],
    'Quantity' : [10,   15,   8,    5,     20]
}

df = pd.DataFrame(data)
print(df)
```

Q4. Indexing using .loc
Using .loc, perform the following:
Display the row with index 10.
Display the Sales value of the row with index 20

```python
# 1. Display the row with index 10
print(df.loc[10])

# 2. Display the Sales value of the row with index 20
print(df.loc[20, 'Sales'])
```

Q5. Indexing using .iloc
Using .iloc, display:
The first 6 rows and first 4 columns.
The Sales value from the 3rd row

```python
# 1. First 6 rows and first 4 columns
print(df.iloc[:6, :4])

# 2. Sales value from the 3rd row (index 2)
print(df.iloc[2, df.columns.get_loc('Sales')])
```

Q6. Row Slicing
Display:
The first 10 rows of the dataset.
Rows 15 to 25

```python
# 1. First 10 rows
print(df.iloc[:10])

# 2. Rows 15 to 25 (inclusive)
print(df.iloc[15:26])
```

Q7. Filtering Data
Filter the dataset to display:
Rows where Region = "East".
Rows where Sales > 1500.

```python
# 1. Rows where Region = 'East'
east = df[df['Region'] == 'East']
print(east)

# 2. Rows where Sales > 1500
high_sales = df[df['Sales'] > 1500]
print(high_sales)
```

Q8. Multiple Condition Filtering
Display rows where:
Region is West
Sales is greater than 1200

```python
# Rows where Region is West AND Sales > 1200
filtered = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]
print(filtered)
```

Q9. Descriptive Statistics
Generate summary statistics using:
describe()
From the result identify:
Mean sales value
Maximum quantity sold
Minimum sales value

```python
# Generate summary statistics
stats = df.describe()
print(stats)

# Identify key values
print("Mean Sales   :", df['Sales'].mean())
print("Max Quantity :", df['Quantity'].max())
print("Min Sales    :", df['Sales'].min())
```

Q10. Aggregation Analysis
Using the Sales column, calculate:
Total sales
Average sales
Maximum sales
Minimum sales


```python
# Sales column aggregations
print("Total Sales   :", df['Sales'].sum())
print("Average Sales :", df['Sales'].mean())
print("Maximum Sales :", df['Sales'].max())
print("Minimum Sales :", df['Sales'].min())
```
